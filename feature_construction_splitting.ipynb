Applynig feature construction

feature construction is get the two feature into 1 columns

```python
x['family_size']=x['SibSp']+x['Parch']+1
```

```python
x.head()
```

```python
def myfunc(num):
  if num==1:
    #alone
    return 0
  elif num > 1 and num <=4:
    #small family
    return 1
  else :
    #large family
    return 2
```

```python
myfunc(4)
```

```python
x['family_type']=x['family_size'].apply(myfunc)
```

```python
x.head()
```

```python
x.drop(columns=['SibSp','Parch','family_size'],inplace=True)
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=20))
```
